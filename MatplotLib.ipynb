```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,20,25,30]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color = 'green' , linestyle = '--',marker = 'o')
plt.title('My Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
x = [1,2,3]
y1 = [1,4,9]
y2 = [1,2,3]
plt.plot(x , y1 , label = 'Squares')
plt.plot(x , y2 , label = 'Linear')
plt.legend()
plt.show()
```

```python
categories = ['A' , 'B' , 'C']
values = [5,7,3]
plt.bar(categories,values)
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,25,30]
plt.scatter(x,y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3],[1,2,3])
plt.subplot(1,2,2)
plt.plot([1,2,3],[3,2,1])
plt.show()
```
