```python
import pandas as pd
df = pd.DataFrame({'key': [1, 2],'Name': ['A', 'B']})
df2 = pd.DataFrame({'key': [1, 2],'Place': ['Kollam', 'Kochi']})
pd.merge(df, df2, on='key')
```

```python
df = pd.DataFrame({'key': [1, 2],'Name': ['A', 'B']})
df2 = pd.DataFrame({'key': [1, 2],'Place': ['Kollam', 'Kochi']})
df.set_index('key').join(df2.set_index('key'))
```
